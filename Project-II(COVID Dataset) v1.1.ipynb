```python
#categorizing data with respect to unique countries

confirmed_by_country = []
death_by_country = [] 
active_by_country = []
recovery_by_country = []
mortality_rate_by_country = [] 

no_cases = []

unique_countries =  list(df_covid['Country'].unique())

all_dates = df_covid["Date"]
last_date = all_dates.iat[-1]
total_confirmed_cases =[]


for i in unique_countries:
    active_cases = df_covid[df_covid["Country"]==i]["Confirmed"].iloc[-1]
    recovered_cases = df_covid[df_covid["Country"]==i]["Recovered"].iloc[-1]
    death_cases = df_covid[df_covid["Country"]==i]["Deaths"].iloc[-1]
    if active_cases> 0:
        confirmed_by_country.append(i)
        active_by_country.append(active_cases)
        death_by_country.append(death_cases) 
        recovery_by_country.append(recovered_cases) 
        total_confirmed = death_cases + recovered_cases + active_cases
        total_confirmed_cases.append(total_confirmed)
              
    else :
        no_cases.append(i)        
        

        
```

```python

country_df = pd.DataFrame({'Country Name': confirmed_by_country, 'Number of Confirmed Cases': total_confirmed_cases,
                          'Number of Deaths': death_by_country, 'Number of Recoveries' : recovery_by_country, 
                          'Number of Active Cases' : active_by_country})
# number of cases per country/region

country_df.style.background_gradient(cmap='Blues')
```

```python
#Sort by confirmed cases

sort_by_country = country_df.sort_values('Number of Confirmed Cases', ascending=False)
sort_by_country.style.background_gradient(cmap='Blues')
```

```python
#creating database for sum of confirmed case, recovered case and deaths in all the countries 

world_cases = []
total_deaths = [] 
recovery_cases = [] 
total_active = [] 
unique_dates = []
total_recovered =[]


dates = df_covid["Date"].unique()

for i in dates:
    death = df_covid[df_covid["Date"]==i]["Deaths"].sum()
    total_deaths.append(death)
    world = df_covid[df_covid["Date"]==i]["Confirmed"].sum()
    world_cases.append(world)
    recovered = df_covid[df_covid["Date"]==i]["Recovered"].sum()
    total_recovered.append(recovered)
    unique_dates.append(i)
    

    
world_cases_df = pd.DataFrame({'Date': dates, 'Number of Confirmed Cases': world_cases,
                          'Number of Deaths': total_deaths, 'Number of Recoveries' : total_recovered, 
                          })


world_cases_df.style.background_gradient(cmap='Blues')

```

```python
#prediction model 
# creating variables for linear regression
#future_forecast created for 20 days in future
days_since_22nd_Jan = np.array([i for i in range(len(dates))]).reshape(-1, 1)
world_cases = np.array(world_cases).reshape(-1, 1)
total_deaths = np.array(total_deaths).reshape(-1, 1)
total_recovered = np.array(total_recovered).reshape(-1, 1)

days_in_future = 20
future_forecast = np.array([i for i in range(len(dates)+days_in_future)]).reshape(-1, 1)
adjusted_dates = future_forecast[:-20]
future_forecast
```

```python
#plottin no. of coronovirus death over time
plt.plot(adjusted_dates, total_deaths)
plt.title('Number of Coronavirus Deaths Over Time', size=30)
plt.xlabel('Days Since 1/22/2020', size=10)
plt.ylabel('Number of Cases', size=10)
plt.xticks(size=10)
plt.yticks(size=10)
plt.show()
```

```python
#plotting no. of world cases over time
adjusted_dates = adjusted_dates.reshape(1, -1)[0]
plt.plot(adjusted_dates, world_cases)
plt.title('Number of Coronavirus Cases Over Time', size=30)
plt.xlabel('Days Since 1/22/2020', size=10)
plt.ylabel('Number of Cases', size=10)
plt.xticks(size=10)
plt.yticks(size=10)
plt.show()
```

```python
#importing libraries
from sklearn.linear_model import LinearRegression
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import PolynomialFeatures
from sklearn.svm import SVR
from sklearn.metrics import mean_squared_error, mean_absolute_error
import datetime
import operator 
```

```python
X_train_confirmed, X_test_confirmed, y_train_confirmed, y_test_confirmed = train_test_split(days_since_22nd_Jan, world_cases, test_size=0.25, shuffle=False)
```

```python
# transform data for polynomial regression
poly = PolynomialFeatures(degree=2)
poly_X_train_confirmed = poly.fit_transform(X_train_confirmed)
poly_X_test_confirmed = poly.fit_transform(X_test_confirmed)
poly_future_forecast = poly.fit_transform(future_forecast)
```

```python
# polynomial regression
linear_model = LinearRegression(normalize=True, fit_intercept=False)
linear_model.fit(poly_X_train_confirmed, y_train_confirmed)
test_linear_pred = linear_model.predict(poly_X_test_confirmed)
linear_pred = linear_model.predict(poly_future_forecast)
```

```python
#plotting the difference between prediction model and test data
plt.plot(y_test_confirmed)
plt.plot(test_linear_pred)
plt.legend(['Test Data', 'Polynomial Regression Predictions'])
#Model prediction is closed to real test data
```

```python
# svm_confirmed = svm_search.best_estimator_
svm_confirmed = SVR(shrinking=True, kernel='poly',gamma=0.01, epsilon=1,degree=5, C=0.1)
svm_confirmed.fit(X_train_confirmed, y_train_confirmed)
svm_pred = svm_confirmed.predict(future_forecast)
```

```python
#plotting real test data and SVM plot prediction
svm_test_pred = svm_confirmed.predict(X_test_confirmed)
plt.plot(y_test_confirmed)
plt.plot(svm_test_pred)
plt.legend(['Test Data', 'SVM Predictions'])
print('MAE:', mean_absolute_error(svm_test_pred, y_test_confirmed))
print('MSE:',mean_squared_error(svm_test_pred, y_test_confirmed))
```

```python
#defining a plot prediction function 
def plot_predictions(x, y, pred, algo_name, color):
    plt.plot(x, y)
    plt.plot(future_forecast, pred, linestyle='dashed', color=color)
    plt.title('Number of Coronavirus Cases Over Time', size=10)
    plt.xlabel('Days Since 1/22/2020', size=10)
    plt.ylabel('Number of Cases', size=10)
    plt.legend(['Confirmed Cases', algo_name], prop={'size': 10})
    plt.xticks(size=10)
    plt.yticks(size=10)
    plt.show()
```

```python
#Plotting data from test and predicted value of linear regression

plot_predictions(adjusted_dates, world_cases, linear_pred, 'Polynomial Regression Predictions', 'red')
#Clearly polynomial regression is showing very good acurracy in prediction
```

```python
plot_predictions(adjusted_dates, world_cases, svm_pred, 'SVM Predictions', 'green')
```
