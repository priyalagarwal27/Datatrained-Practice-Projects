```python
#Using Linear Regression
from sklearn.linear_model import LinearRegression
X = df_score.drop("Happiness Score", axis = 1)
lm = LinearRegression()
lm.fit(X, df_score["Happiness Score"])
```

```python
print("Estimated Intercept ", lm.intercept_)
```

```python
print("Number of coefficients ", lm.coef_)
```

```python
coef = zip(X.columns, lm.coef_)
coef_df = pd.DataFrame(list(zip(X.columns, lm.coef_)), columns=['features', 'coefficients'])
coef_df
```

```python
lm.predict(X)[0:100]
```

```python
sns.scatterplot(x = lm.predict(X),y = df_score["Happiness Score"], data=df_score)
```

```python
msehappy = np.mean(( df_score["Happiness Score"]- lm.predict(X)) ** 2 ) 
print(msehappy)
```

```python
lm2=LinearRegression()
lm2.fit(X[['Family']],df_score["Happiness Score"] )
```

```python
msefamily = np.mean((df_score["Happiness Score"] - lm2.predict(X[['Family']])) **2)
print(msefamily)
```
