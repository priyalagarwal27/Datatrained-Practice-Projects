```python
#Impute the null values 
impute=["Age","Embarked"]
for col in impute:
    if df_titanic[col].dtype!="object":
        df_titanic[col].fillna(df_titanic[col].mean(),inplace=True)
    else:
        df_titanic[col].fillna(df_titanic[col].mode()[0],inplace=True)
```

```python
#Cabin Class ,Name , Pid column will do nothing in prediction so we will drop it 
df_titanic.drop(columns=["Passangerid","Name","Cabin"],inplace=True)
```

```python
#using pd.get_dummies function to convert categorical columns into numeric form 
df_titanic_final=pd.get_dummies(df_titanic,drop_first=True)
```

```python
df_titanic_final
```

```python
#Seperate input and output variables
df_x=df_titanic_final.drop(columns=["Pclass"])
y=df_titanic_final[["Pclass"]]
```

```python
#Using decision tree regressor
from sklearn.metrics import mean_absolute_error
from sklearn.metrics import mean_squared_error
from sklearn.metrics import r2_score
from sklearn.model_selection import train_test_split
from sklearn.tree import DecisionTreeRegressor
def maxr2_score(regr,df_x,y):
    max_r_score=0
    for r_state in range(42,100):
        x_train, x_test, y_train, y_test = train_test_split(df_x, y,random_state = r_state,test_size=0.20)
        regr.fit(x_train,y_train)
        y_pred = regr.predict(x_test)
        r2_scr=r2_score(y_test,y_pred)
        print("r2 score corresponding to ",r_state," is ",r2_scr)
        if r2_scr>max_r_score:
            max_r_score=r2_scr
            final_r_state=r_state
    print("max r2 score corresponding to ",final_r_state," is ",max_r_score)
    return final_r_state
```

```python
regr=DecisionTreeRegressor()
final_r_state=maxr2_score(regr,df_x,y)
```

```python
#Using another model KNN regressor for that we have to scale our input set 
from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
x = sc.fit_transform(df_x)
x=pd.DataFrame(x,columns=df_x.columns)
```

```python
#using grid search
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import GridSearchCV
parameters = {'n_neighbors':range(2,10)}
knr = KNeighborsRegressor()
clf = GridSearchCV(knr, parameters, cv=5)
clf.fit(x,y)
clf.best_params_
```

```python
knr=KNeighborsRegressor(n_neighbors=7)
final_r_state=maxr2_score(knr,x,y)
```

```python
#Linear Regression
from sklearn.linear_model import LinearRegression
lr=LinearRegression()
final_r_state=maxr2_score(lr,df_x,y)
```
